```python
'''
  Ejercicio 3001
'''
class Galleta:
    def __init__(self, nombre, forma):
        self.nombre = nombre
        self.forma = forma

    def __hornear__(self):
        print(f"Esta galleta de {self.nombre} ha sido horneada en forma de {self.forma}. \nBuen provecho.")

galleta = Galleta("Chispas de chocolate", "estrella")

galleta.__hornear__()
```

```python
'''
  Ejercicio 3003
'''
class Nota:

  def __init__(self, nota, nombreEstudiante):
    self.nombreEstudiante = nombreEstudiante
    self.nota = nota

  def ha_pasado(self):
    print(f"El estudiante {self.nombreEstudiante}", "ha aprobado. " if self.nota > 75 else "ha suspendido.")

nota = Nota(34, "Pedro")
nota.ha_pasado()

nota2 = Nota(90, "Juan")
nota2.ha_pasado()
```

```python
'''
  Ejercicio 3004
'''
import math as m
class Circulo:

  def __init__(self, x, y, r):
    self.x = x
    self.y = y
    self.r = r

  def area(self):
    print(f"El area del circulo es: {(m.pi * (self.r ** 2)):.2f}")

  def perimetro(self):
    print(f"El perimetro del circulo es: {(m.pi * (2 * self.r)):.2f}")

  def mostrar_propiedades(self):
    print(f"La posición del centro del circulo es: ({self.x}, {self.y}) y su radio es: {self.r}")

c1 = Circulo(2, 4, 5)
c1.area()
c1.perimetro()
c1.mostrar_propiedades()
```

```python
'''
  Ejercicio 3005
'''
class Operacion:

  def __init__(self, x, y):
    self.x = x
    self.y = y

  def suma(self):
    print(f"La suma de {self.x} + {self.y} es: {self.x + self.y}")

  def multiplicacion(self):
    print(f"El resultado de la multiplicación de {self.x} x {self.y} es: {self.x * self.y}")

  def division(self):
    print(f"El resultado de la división de {self.x} / {self.y} es: {self.x / self.y}")

op = Operacion(2, 5)
op.suma()
op.multiplicacion()
op.division()
```

```python
'''
  Ejercicio 3006
'''

class Persona:

  def __init__(self,nombre,edad,genero):
    self.nombre = nombre
    self.edad = edad
    self.genero = genero

  def presentarse(self):
    return f"Hola soy {self.nombre}, tengo {self.edad} años y soy del genero {self.genero}. "

  def esAdulto(self):
    return True if self.edad >= 18 else False


p = Persona("Juan", 18, "Masculino")
print(p.presentarse())
print(p.esAdulto())
```

```python
'''
  Ejercicio 3009
'''
class Vehiculo:
  def __init__(self, marca):
    self.marca = marca
    self.velocidadInicial = 0

  def acelerar(self, v):
    self.velocidadInicial += v

  def desacelerar(self,v):
    self.velocidadInicial -= v

  def mostrar_velocidad(self):
    print(f"La velocidad actual del vehiculo con marca {self.marca} es de {self.velocidadInicial} Km/h. ")

class Coche(Vehiculo):
  def __init__(self, marca):
    super().__init__(marca)

  def Tocar_Calaxon(self):
    print("PIIII PIIIIIII ")

c = Coche("Toyota")
c.acelerar(100)
c.mostrar_velocidad()
c.desacelerar(23)
c.mostrar_velocidad()
c.Tocar_Calaxon()
```

```python
'''
  Ejercicio 3010
'''

class Empleado:
  def __init__(self,nombre,funcion,salario):
    self.nombre = nombre
    self.funcion = funcion
    self.salario = salario
    self.numHorasTrabajadas = 0

  def trabajar(self,num_horas):
    self.numHorasTrabajadas += num_horas
    return f"El número de horas trabajadas hasta el momento es: {self.numHorasTrabajadas} "

  def info_sueldo(self):
    return f"El sueldo del empleado es: {self.salario} € "

  def dar_aumento(self,cantidad):
    self.salario += cantidad

  def info_funcion(self):
    return f"La función del empleado es: {self.funcion} "

emp = Empleado("Juan", "Programador", 1000)
print(emp.trabajar(10))
print(emp.info_sueldo())
emp.dar_aumento(45)
print(emp.info_sueldo())
print(emp.info_funcion())
```

```python
'''
    Ejercicio 3011
'''

class Claculo_numerico:
    def __init__(self, numero):
        self.numero = numero

    def calculo_factorial(self):
        result = 1
        for i in range(1, self.numero + 1):
            result *= i
        return result

    def lista_divisores(self):
        lista = []
        for i in range(1, self.numero):
            if self.numero % i == 0:
                lista.append(i)

        return lista

    def esPrimo(self):
        for i in range(2, int(self.numero**0.5) + 1):
            if self.numero % i == 0:
                return f"El número {self.numero} no es primo. "
        return f"El número {self.numero} es primo. "

    def esPar(self):
        if self.numero % 2 == 0:
            return f"El número {self.numero} es par. "
        return f"El número {self.numero} no es par. "

c = Claculo_numerico(15)
print(c.calculo_factorial())
print(c.lista_divisores())
print(c.esPrimo())
print(c.esPar())
```

```python
'''
  Ejercicio 3012
'''
class Video:

    def __init__(self, titulo, duracion, categoria):
        self.titulo = titulo
        self.duracion = duracion
        self.categoria = categoria

    def mirar_video(self):
        print("Estas viendo el video. ")

    def detener_video(self):
        print("Has detenido el video. ")

class Audio:

    def __init__(self, titulo, nombreArtista):
        self.titulo = titulo
        self.nombreArtista = nombreArtista

    def escuchar_audio(self):
        print("Estas escuchando el audio. ")

    def detener_audio(self):
        print("Has detenido el audio. ")

class Media(Video, Audio):

    def __init__(self, titulo, duracion, categoria, nombreArtista):
        Video.__init__(self, titulo, duracion, categoria)
        Audio.__init__(self, titulo, nombreArtista)

m1 = Media("Concierto Live", "60", "Trap", "Bad Bunny")
m1.mirar_video()
m1.escuchar_audio()
m1.detener_video()
m1.detener_audio()
```
